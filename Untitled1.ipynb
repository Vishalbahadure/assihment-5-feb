```python
Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.
Class:
A class is a blueprint or template for creating objects. It defines a set of attributes (data members) and 
methods (functions) that the objects created from the class will have. Think of a class as a blueprint for a specific type of object.

Object:
An object is an instance of a class. It represents a real-world entity or concept and encapsulates data (attributes) and
behavior (methods) related to that entity.
Objects are created from a class and can interact with other objects and the program itself.


class Car:
    
    def __init__(self, make, model, year):
        self.make = make         
        self.model = model        
        self.year = year          
        self.is_running = False   

    def start(self):
        self.is_running = True
        print(f"{self.year} {self.make} {self.model} is now running.")

    def stop(self):
        self.is_running = False
        print(f"{self.year} {self.make} {self.model} has stopped.")

car1 = Car("Toyota", "Camry", 2022)
car2 = Car("Honda", "Civic", 2021)

print(f"{car1.year} {car1.make} {car1.model}") 
car1.start()  
car1.stop() 

print(f"{car2.year} {car2.make} {car2.model}")  # Output: 2021 Honda Civic
car2.start()  # Output: 2021 Honda Civic is now running.
car2.stop()   # Output: 2021 Honda Civic has stopped.
```

```python
Q2. Name the four pillars of OOPs.

The four pillars of object-oriented programming (OOP) are:

Encapsulation: Encapsulation is the concept of bundling data (attributes or properties) and the methods (functions) that operate on that data into a single unit called a class. It allows you to control access to the internal state of an object by defining access modifiers (public, private, protected), ensuring that data is accessed and modified in a controlled way.

Inheritance: Inheritance is a mechanism that allows a new class (subclass or derived class) to inherit properties and behaviors (attributes and methods) from an existing class (superclass or base class). It promotes code reuse and the creation of a hierarchy of classes, where subclasses can extend or override the functionality of their parent classes.

Polymorphism: Polymorphism allows objects of different classes to be treated as objects of a common superclass. It allows you to write code that can work with objects of various types without needing to know their specific types at compile time. Polymorphism is typically achieved through method overriding and interfaces (in languages like Java) and promotes flexibility and extensibility.

Abstraction: Abstraction is the process of simplifying complex reality by modeling classes based on the essential properties and behaviors while hiding unnecessary details. It allows you to create abstract classes and methods that define a blueprint for derived classes. Abstraction helps in managing complexity and focusing on what an object does rather than how it does it.

```

```python
Q4. Why self is used in OOPs?
Differentiating Instance Variables: Within a class, self is used to distinguish instance variables (attributes that belong to a specific object) from local variables (variables that are local to a specific method). This allows you to work with the object's own data.

Accessing Object Attributes: self is used to access the attributes of an object. For example, if you have an attribute called name in a class, you would access it as self.name within the class's methods.

Modifying Object Attributes: self is used to modify the attributes of an object. You can assign new values to attributes using self.attribute_name.

Invoking Object Methods: self is used to call methods on the object itself. For example, if you have a method called display_info in a class, you would call it as self.display_info() within the class.

```

```python
Q5. What is inheritance? Give an example for each type of inheritance.

Inheritance is a fundamental concept in object-oriented programming (OOP) that allows you to create a new class (subclass or derived class) by inheriting properties and behaviors (attributes and methods) from an existing class (superclass or base class). Inheritance promotes code reuse and the creation of a hierarchy of classes, where subclasses can extend or override the functionality of their parent classes.

There are several types of inheritance, including single inheritance, multiple inheritance, hierarchical inheritance, and multilevel inheritance. Let's provide an example for each of these types:
ingle Inheritance:

Single inheritance occurs when a subclass inherits from a single superclass.
python
Copy code
# Single Inheritance Example
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

dog = Dog()
print(dog.speak())  # Output: "Woof!"
Multiple Inheritance:

Multiple inheritance occurs when a subclass inherits from two or more superclasses.
python
Copy code
# Multiple Inheritance Example
class Parent1:
    def method1(self):
        pass

class Parent2:
    def method2(self):
        pass

class Child(Parent1, Parent2):
    def child_method(self):
        pass

child = Child()
child.method1()  # Access method from Parent1
child.method2()  # Access method from Parent2
Multilevel Inheritance:

Multilevel inheritance occurs when a subclass inherits from a superclass, and another subclass inherits from that subclass.
python
Copy code
# Multilevel Inheritance Example
class Grandparent:
    def method1(self):
        pass

class Parent(Grandparent):
    def method2(self):
        pass

class Child(Parent):
    def child_method(self):
        pass

child = Child()
child.method1()  # Access method from Grandparent
child.method2()  # Access method from Parent
```
